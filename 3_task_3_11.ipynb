11.Apply and explore various plotting functions on Pima Indians Diabetes data set for performing the
following:
d) Normal values
e) Density and contour plots
f) Three-dimensional plotting



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px

# Load the Pima Indians Diabetes dataset
# You may need to download the dataset or replace the path with your dataset location
df = pd.read_csv('/content/diabetes.csv')

# Display normal values using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title('Box Plot of Glucose Levels for Diabetes Outcome')
plt.xlabel('Outcome (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Glucose Level')
plt.show()

# Density and contour plots
plt.figure(figsize=(12, 8))
sns.kdeplot(x='Glucose', y='BMI', data=df, fill=True, cmap='viridis', levels=20)
plt.title('Density and Contour Plot of Glucose vs BMI')
plt.xlabel('Glucose Level')
plt.ylabel('BMI')
plt.show()

# Three-dimensional plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Pregnancies'], df['Glucose'], df['BMI'], c=df['Outcome'], cmap='viridis', s=50)
ax.set_xlabel('Pregnancies')
ax.set_ylabel('Glucose Level')
ax.set_zlabel('BMI')
ax.set_title('3D Scatter Plot of Pregnancies, Glucose, and BMI')
plt.show()

# Interactive 3D plot using Plotly Express
fig = px.scatter_3d(df, x='Pregnancies', y='Glucose', z='BMI', color='Outcome',
                    size_max=10, opacity=0.7, title='Interactive 3D Scatter Plot')
fig.update_layout(scene=dict(xaxis_title='Pregnancies', yaxis_title='Glucose', zaxis_title='BMI'))
fig.show()
```
