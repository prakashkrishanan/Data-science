12.Apply and explore various plotting functions on Pima Indians Diabetes data set for performing the
following:
g) Correlation and scatter plots
h) Histograms
i) Three-dimensional plotting



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Pima Indians Diabetes dataset (replace 'path_to_dataset.csv' with your actual path)
data = pd.read_csv('/content/drive/MyDrive/diabetes.csv')

# Display the first few rows of the dataset
print("Pima Indians Diabetes dataset:")
print(data.head())

# g) Correlation and scatter plots
correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Scatter plots
sns.pairplot(data, hue='Outcome', markers=['o', 's'], palette='husl')
plt.suptitle("Pairplot of the Pima Indians Diabetes dataset", y=1.02)
plt.show()

# h) Histograms
data.hist(figsize=(12, 8), bins=20)
plt.suptitle("Histograms of Pima Indians Diabetes dataset", y=1.02)
plt.show()

# i) Three-dimensional plotting
# Note: For three-dimensional plotting, we'll use a subset of the data for simplicity.
subset_data = data[['Glucose', 'BMI', 'Age']]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(subset_data['Glucose'], subset_data['BMI'], subset_data['Age'], c=data['Outcome'], cmap='viridis')
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
ax.set_title('Three-dimensional Plot of Glucose, BMI, and Age')
plt.show()
```
